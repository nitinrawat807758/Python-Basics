```python
#Write a function to demonstrate the behavior of mutable and immutable arguments.
def modify_arguments(immutable_arg, mutable_arg):
    # Try to modify the immutable argument
    immutable_arg += 10
    print("Inside function (immutable):", immutable_arg)

    # Modify the mutable argument
    mutable_arg.append(4)
    print("Inside function (mutable):", mutable_arg)

# Immutable type: int
x = 5

# Mutable type: list
y = [1, 2, 3]

print("Before function call:")
print("Immutable x:", x)
print("Mutable y:", y)

modify_arguments(x, y)

print("\nAfter function call:")
print("Immutable x:", x)   # Will remain unchanged
print("Mutable y:", y)     # Will be changed
```

```python
# Write a program that performs basic arithmetic operations on two user-input numbers.
def main():
    print("Basic Arithmetic Operations")

    # Get user input and convert to float
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))

    # Perform operations
    addition = num1 + num2
    subtraction = num1 - num2
    multiplication = num1 * num2

    # Handle division by zero
    if num2 != 0:
        division = num1 / num2
    else:
        division = "Undefined (cannot divide by zero)"

    # Display results
    print("\nResults:")
    print(f"{num1} + {num2} = {addition}")
    print(f"{num1} - {num2} = {subtraction}")
    print(f"{num1} * {num2} = {multiplication}")
    print(f"{num1} / {num2} = {division}")

if __name__ == "__main__":
    main()
```

```python
# Write a Python program to convert user input from string to integer, float, and boolean types.
def convert_input(user_input):
    # Try converting to integer
    try:
        int_value = int(user_input)
    except ValueError:
        int_value = None

    # Try converting to float
    try:
        float_value = float(user_input)
    except ValueError:
        float_value = None

    # Convert to boolean (basic interpretation)
    true_values = ['true', '1', 'yes', 'on']
    false_values = ['false', '0', 'no', 'off']
    user_input_lower = user_input.strip().lower()
    if user_input_lower in true_values:
        bool_value = True
    elif user_input_lower in false_values:
        bool_value = False
    else:
        bool_value = None

    return int_value, float_value, bool_value

# Main program
user_input = input("Enter a value: ")
int_val, float_val, bool_val = convert_input(user_input)

print(f"Integer conversion: {int_val}")
print(f"Float conversion: {float_val}")
print(f"Boolean conversion: {bool_val}")
```

```python
# Write code to demonstrate type casting with list elements.
# Sample list of strings
string_list = ['10', '20.5', '30', 'true', 'False', '0']

# Convert string list elements to various types
int_list = []
float_list = []
bool_list = []

for item in string_list:
    # Convert to int if possible
    try:
        int_list.append(int(float(item)))  # Handles both int and float strings
    except ValueError:
        int_list.append(None)

    # Convert to float if possible
    try:
        float_list.append(float(item))
    except ValueError:
        float_list.append(None)

    # Convert to bool using a simple mapping
    lower_item = item.lower()
    if lower_item in ['true', '1', 'yes', 'on']:
        bool_list.append(True)
    elif lower_item in ['false', '0', 'no', 'off']:
        bool_list.append(False)
    else:
        bool_list.append(None)

# Display the results
print("Original list:", string_list)
print("Converted to int:", int_list)
print("Converted to float:", float_list)
print("Converted to bool:", bool_list)
```

```python
# Write a Python program to find the sum of all even numbers between 1 and 50.
# Initialize sum
even_sum = 0

# Loop through numbers from 1 to 50
for i in range(1, 51):
    if i % 2 == 0:
        even_sum += i

# Display the result
print("The sum of all even numbers from 1 to 50 is:", even_sum)
```

```python
# Write a program to reverse a string using a while loop.
# Get input from the user
original_string = input("Enter a string: ")

# Initialize variables
reversed_string = ""
index = len(original_string) - 1

# Use while loop to reverse the string
while index >= 0:
    reversed_string += original_string[index]
    index -= 1

# Display the reversed string
print("Reversed string:", reversed_string)

```
